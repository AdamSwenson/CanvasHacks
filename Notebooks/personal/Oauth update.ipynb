```python
# ============================
# 2. Create XOAUTH2 String
# ============================
def make_auth_bytes(fromAddr, access_token):
    auth_string = f"user={fromAddr}\x01auth=Bearer {access_token}\x01\x01"
    return base64.b64encode(auth_string.encode()).decode()

auth_bytes = make_auth_bytes(user, access_token)

content="Hello, \r\n This is a sample email"

message=MIMEText(content)
message['From']=user
message['To']=destination
message['Subject']='Test Email from SMTP Oauth App' 

with smtplib.SMTP(smtp_server, smtp_port) as server:
    server.set_debuglevel( True )
    server.ehlo()
    server.starttls()
    server.ehlo()
    code, resp = server.docmd("AUTH", "XOAUTH2 " + auth_bytes)
    if code > 400:
        print('borp')
    print("Debug, at server.docmd")
    print("AUTH response:", code, resp)
    server.sendmail(fromAddr, toAddr, message.as_string())
    server.close()
```
